```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = pd.read_csv('./data/iris.csv')
```

```python
iris.plot(kind = 'scatter',x = 'sepal_length',y ='sepal_width')
```

```python
plt.figure(figsize = (10,10))
sns.set_style("whitegrid")
sns.FacetGrid(iris,hue='species',height=10).map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',diag_kind='kde')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris,height=10,hue='species').map(sns.distplot,'petal_length').add_legend()
plt.plot()
```
